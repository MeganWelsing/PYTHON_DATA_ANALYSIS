## Replacing null values

```python
for column in data.columns:
    if column != 'ID':
       data[column].fillna(data[column].mean(), inplace=True)
```

## Method used for replacing null values: Using mean values of each column

### Reason: When null values are encountered in a data set, there are many methods used to tackle that issue. One common method is deleting rows. However, this method leads to loss of information and can produce inaccurate results. Replacing null values with the mean values of each column is a better way of handling null values in a dataset. This is an approximation that adds variance to the data set. This method prevents data loss and promotes precision in data analysis.

## Showing hidden inferences in the dataset

### Z-scores are a statistical measure used to determine how far away a particular data point is from the mean of the dataset in terms of standard deviations. A Z-score indicates the number of standard deviations away from the mean that a particular data point falls.

### An outlier is a data point that falls far outside the range of the other data points in a dataset. Outliers can be identified using various methods, such as using Z-scores or box plots.

### Z-scores can be used to identify outliers in a dataset by setting a threshold for the number of standard deviations away from the mean that a data point should fall within. Data points that fall outside this threshold can be considered outliers.

### Outliers can have a significant impact on statistical analysis, as they can skew the results and lead to incorrect conclusions. Therefore, it is important to identify and handle outliers appropriately when analyzing data.

```python
for column in data.columns:
    if column != 'ID' :
    #Calculating the z-scores for each data point of each column
       z_scores = (data[column] - data[column].mean()) / data[column].std()
        
    #output below is all z_scores values for each column
       print(z_scores)

     #outliers are identified using a threshold of 3 standard deviations
       outliers = data[abs(z_scores) > 3]
```
